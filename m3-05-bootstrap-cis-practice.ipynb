```python
import numpy as np
import pandas as pd
import scipy.stats as st

# Reuse bootstrap_ci from Task 2
def bootstrap_ci(data, stat_func=np.mean, n_boot=10_000, ci_level=0.95):
    data = np.asarray(data)
    n = len(data)
    boot_stats = np.empty(n_boot)
    for i in range(n_boot):
        sample = np.random.choice(data, size=n, replace=True)
        boot_stats[i] = stat_func(sample)
    alpha = 1 - ci_level
    lower = np.percentile(boot_stats, 100 * (alpha / 2))
    upper = np.percentile(boot_stats, 100 * (1 - alpha / 2))
    return lower, upper

# Generate synthetic data
np.random.seed(42)
n_rows = 500
data = pd.DataFrame({
    'score': np.random.normal(75, 10, n_rows),
    'passed': np.random.binomial(1, 0.7, n_rows)
})

# Bootstrap CIs
ci_mean_score = bootstrap_ci(data['score'])
ci_prop_passed = bootstrap_ci(data['passed'])

# Normal-Approximation CIs
n = len(data)

# Mean
mean_score = data['score'].mean()
se_score = data['score'].std(ddof=1) / np.sqrt(n)
normal_ci_mean = st.t.interval(0.95, df=n-1, loc=mean_score, scale=se_score)

# Proportion
p_hat = data['passed'].mean()
se_prop = np.sqrt(p_hat * (1 - p_hat) / n)
z = 1.96
normal_ci_prop = (p_hat - z*se_prop, p_hat + z*se_prop)

# Comparison table
comparison_df = pd.DataFrame({
    'Statistic': ['Mean', 'Proportion'],
    'Bootstrap Lower': [ci_mean_score[0], ci_prop_passed[0]],
    'Bootstrap Upper': [ci_mean_score[1], ci_prop_passed[1]],
    'Normal Lower': [normal_ci_mean[0], normal_ci_prop[0]],
    'Normal Upper': [normal_ci_mean[1], normal_ci_prop[1]]
})

print(comparison_df)
```

Are the two approaches giving similar intervals?
For the mean (score), the bootstrap CI and the normal-approximation CI are very similar. This is because the sample size is large and the mean is approximately normally distributed (Central Limit Theorem).
For the proportion (passed), the intervals are also quite close, although small differences may appear if the proportion is very close to 0 or 1.
Where do they diverge?

The divergence can appear for small sample sizes, skewed distributions, or for statistics that are not normally distributed, such as the median or other quantiles.

Normal-approximation methods assume symmetry and normality of the estimator; bootstrap does not, so it adapts to the actual shape of the sampling distribution.


For which statistic is the bootstrap approach especially useful, and why?

Bootstrap is particularly useful for medians, percentiles, or any non-standard statistic where the normal approximation is unreliable.

Example: the median may have a skewed or discrete sampling distribution; a normal-based CI could be misleading, while bootstrap directly uses the empirical distribution of resampled statistics to produce more accurate intervals.
